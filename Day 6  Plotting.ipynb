```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
output_notebook()
```

```python
x = np.random.random(100)
y = np.random.random(100)

p = figure(tools='pan,box_zoom,reset,save,crosshair,hover', title='random', x_axis_label='x', y_axis_label='y')
p.circle(x,y)
show(p)
```
